```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
final=pd.DataFrame()
for j in range(1,39):
  webpage=requests.get('https://www.airlinequality.com/airline-reviews/british-airways/page/{}/?sortby=post_date%3ADesc&pagesize=100'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('article',itemprop="review")
  name=[]
  rating=[] 
  varified=[]
  reviews=[]
  recommand=[]
  published=[]
  country=[]
    

  for i in company:

    try:
       rating.append(i.find("div", class_ = "rating-10").span.text)
    except:
       rating.append(np.nan)

    try:
       name.append(i.find('span',itemprop="name").text.strip())
    except:
       name.append(np.nan)

#     try:
#        rating.append(i.find('div',itemprop="reviewRating").text.strip())
#     except:
#        rating.append(np.nan)
   
    try:

      varified.append(i.find('em').text.strip())
    except:
      varified.append(np.nan)

   
    try:

      reviews.append(i.find('div',class_="text_content").text.strip())
    except:
      reviews.append(np.nan)
    try:

      recommand.append(i.find('td',class_="review-value rating-no").text.strip())
    except:
      recommand.append(np.nan)
    try:

      country.append(i.find('h3').span.next_sibling.text.strip(' ()'))
    except:
      country.append(np.nan)
    try:

      published.append(i.find('time').text.strip())
    except:
      published.append(np.nan)
    
    
    

  df=pd.DataFrame({'name':name,
    'rating':rating,
    'varified':varified,
    'reviews':reviews,
    'recommanded':recommand,
    'country':country,
    'published':published
    })
  
  final=final.append(df,ignore_index=True)



```
